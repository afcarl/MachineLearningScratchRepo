```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#data fields (data taken from https://www.ncdc.noaa.gov/cdo-web/):
#DATE - YYYYMMDD
#EMXP - Extreme maximum daily precipitation (tenths of mm)
#MMNT - Monthly Mean minimum temperature (tenths(!) of degrees Celsius)
#EMXT - Extreme maximum daily temperature (tenths(!) of degrees Celsius)
#MXSD - Maximum snow depth (mm)
#MMXT - Monthly Mean maximum temperature (tenths(!) of degrees Celsius)
#TPCP - Total precipitation (tenths of mm)
#EMNT - Extreme minimum daily temperature (tenths(!) of degrees Celsius)
#MNTM - Monthly mean temperature (tenths(!) of degrees Celsius)
#TSNW - Total snow fall (mm)

#datacity = pd.read_csv("TorontoWeatherCity.csv")      #Date range: 1840-present, but tons of missing data!
datarawRH = pd.read_csv("TorontoWeatherRH.csv")        #Date range: 1959-present, Richmond Hill
print(datarawRH.columns)
```

```python
#data prep - tenths of degrees to degrees, convert date to a continuous numerical scale
dataRH = datarawRH
dataRH["DATE"] = datarawRH["DATE"]/10000 - 0.0101
dataRH["DATE"] = (dataRH["DATE"] - dataRH["DATE"].round())*8.32 + dataRH["DATE"].round()
col = dataRH.columns
for i in xrange(3,len(col)):
    dataRH[col[i]] = datarawRH[col[i]]/10
```

```python
#new columns, data engineering
dataRH["MMDiff"] = dataRH["MMXT"] - dataRH["MMNT"]    #max - min of mean monthly temp
dataRH["EMDiff"] = dataRH["EMXT"] - dataRH["EMNT"]    #max - min of extreme monthly temp

#plot
plt.plot(dataRH["DATE"],dataRH["MNTM"],label="Mean Temp.")
plt.plot(dataRH["DATE"],dataRH["MMDiff"],label="Max - Min Average")
plt.plot(dataRH["DATE"],dataRH["EMDiff"],label="Max - Min Extreme")
plt.legend(loc="upper left", fontsize = 8)
plt.title("Raw Data plotted")
plt.ylabel("Temperature (Celsius)")
plt.xlabel("Year")
#plt.xlim([1990,2008])
```
