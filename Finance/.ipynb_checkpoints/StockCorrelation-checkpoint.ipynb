### By: Ari Silburt
In the management of a financial portfolio one important consideration is the relationship between the portfolio's various stocks. For example, in order to efficiently minimize a risk it is important to diversify the stocks in a portfolio. Thus, if one collection of stocks are suffering, the entire portfolio does not neccessarily suffer.

For this analysis we will make use of an unsupervised machine learning technique called agglomerate hierarchical clustering. This technique looks for clusters in a dataset without needing to know exactly what the meaning of these clusters are. As we will see, certain sectors are naturally more diversified than others. 

```python
import datetime
from pandas_datareader import data as web
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
```

```python
#Get data
dataname = "S&PIT"     #options are DJIA, S&PEnergy, S&PIT
companies = pd.read_csv("tickers_"+dataname+".csv")
companies["Ticker"] = companies["Tickers"]
companies = companies.set_index("Tickers")
start = datetime.datetime(2015, 1, 1)
end = datetime.datetime.today()
data = web.DataReader(companies["Ticker"], 'yahoo', start, end);
list(data)
```

```python
#Isolate Adjusted Close values (typically used for correlations), remove stocks with large streaks of NaN
df_raw = data["Adj Close"]
isnull = df_raw.isnull().sum()
for ticker in companies["Ticker"]:
    if isnull[ticker] > 50:
        df_raw.drop(ticker, axis=1, inplace=True)
```

```python
df_raw.head()
```

```python
#A messy plot of all the adjusted closing prices
%matplotlib inline
plt.plot(df_raw);
```

### Processing Data
Processing and preparation of the data is arguably more important than the analysis itself. Here we will focus on three processing techniques:
1. Scale Data
2. Smooth Data
3. Use derivative

```python
df = df_raw.copy()

#Scale Data
df = (df - df.median())/df.std()
dfscaled = df.copy()

#Smooth Data - 
smooth = 1
if smooth == 1:
    window = 7
    df = df.rolling(window=window).mean()
    df = df[(window-1):]

#literature says that using derivative is much better.
diff = 1
if diff == 1:
    df = df.diff(axis=0)
    df = df[1:]
```

### Data Analysis
#### Distance Measure for clustering
In order to cluster the stocks we need an inherest distance measurement. The most popular one which we will use is the standard correlation coefficient:

<h3 align="center"> $\rho(x,y) = \frac{cov(x,y)}{\sigma_x \sigma_y} $ </h3>

Other distance measures can be used, like Maximal Information Compression Index, Least Square Regression Error, etc.

```python
#find correlation matrix, i.e. the "distances" between each stock
corr = df.corr()
size = 7
fig, ax = plt.subplots(figsize=(size, size))
ax.matshow(corr,cmap=cm.get_cmap('coolwarm'), vmin=0,vmax=1)
plt.xticks(range(len(corr.columns)), corr.columns);
plt.yticks(range(len(corr.columns)), corr.columns);
```

#### Clusters of Correlation - Agglomerate
The next step is to look for clusters of correlations using the agglomerate hierarchical clustering technique. Its primary advantage over other clustering methods is that you don't need to guess in advance how many clusters there might be. Agglomerate Clustering first assigns each data point into its own cluster, and gradually merges clusters until only one remains. It's then up to the user to decide how many clusters are present.

Linkage does the actual clustering in one line of code, and returns a list of the clusters joined in the format: Z=[idx1, idx2, distance, sample_count]

There are also different options for the measurement of the distance. The option we will choose is the average distance measurement, but others are possible (ward, single, centroid, etc.). 

```python
from scipy.cluster.hierarchy import dendrogram, linkage
Z = linkage(corr, 'average')
Z[0]
```

#### Cophenetic Correlation coefficient
It's important to get a sense of how well the clustering performs. One measure is the Cophenetic Correlation Coefficient, $c$. This compares (correlates) the actual pairwise distances of all your samples to those implied by the hierarchical clustering. The closer $c$ is to 1, the better the clustering preserves the original distances. Generally  $c$ > 0.75 is consistered a good cluster fit. Of course, other accuracy checks are possible.

```python
from scipy.cluster.hierarchy import cophenet
from scipy.spatial.distance import pdist
import pylab
c, coph_dists = cophenet(Z, pdist(corr))
c
```

#### Dendogram
The best way to visualize an agglomerate clustering algorithm is through a dendogram, which displays a cluster tree, the leaves being the individual stocks and the root being the final single cluster. The "distance" between each cluster is shown on the y-axis, and thus the longer the branches are, the less correlated two clusters are.

```python
plt.figure(figsize=(25, 10))
labelsize=20
ticksize=15
plt.title('Hierarchical Clustering Dendrogram for '+dataname, fontsize=labelsize)
plt.xlabel('stock index', fontsize=labelsize)
plt.ylabel('distance', fontsize=labelsize)
dendrogram(
    Z,
    leaf_rotation=90.,  # rotates the x axis labels
    leaf_font_size=8.,  # font size for the x axis labels
    labels = corr.columns
)
pylab.yticks(fontsize=ticksize)
pylab.xticks(rotation=-90, fontsize=ticksize)
plt.savefig('dendogram_'+dataname+'.pdf')
plt.show()
```

```python
#Plot a sample correlation
#sA = "GS"
#sB = "JPM"
sA = companies["Ticker"].iloc[0]
sB = companies["Ticker"].iloc[1]
plt.plot(dfscaled[sA],label=companies["Name"].ix[sA]+', '+sA)
plt.plot(dfscaled[sB],label=companies["Name"].ix[sB]+', '+sB)
plt.title('Stock Correlation = '+str(corr[sA][sB]))
plt.ylabel('Scaled Time Series')
plt.legend(loc='upper left',prop={'size':10})
```

### Rolling Correlations
For the correlations we found above, we've only calculated the average correlation between stocks over a long period of time. Of course, these correlations can vary over time. One way to investigate this is using a rolling correlation.

As usual, this can be done magically using pandas. For all stocks you can simultaneously calculate a rolling (trailing) correlation in one line of code.  

```python
window = 100
correls = df.rolling(window=100).corr()
correls = correls[(window-1):,:,:]    #remove NaN values at the beginning
```

```python
#plot a sample rolling correlation
correls.loc[:,sA,sB].plot()
correls.loc[:,sA,sB].mean(), correls.loc[:,sA,sB].std()
```

## Some Conclusions
If you try this analysis for all three datasets (DJIA, S&PEnergy, S&PIT) and look at the resulting dendograms, you will notice that stocks within the Energy sector (S&PEnergy) are highly correlated to each other, while the stocks within the tech sector (S&PIT) are not very correlated. This intuitively makes sense. Companies dependent on a single commodity (e.g. oil) will change together if the underlying commodity changes, while companies grouped together that don't actually have that much in common (e.g. the tech sector) will not change together in the market.
